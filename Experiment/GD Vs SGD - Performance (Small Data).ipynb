```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import gc
import warnings
warnings.filterwarnings('ignore')
```

```python
eta = 0.1
X = 2 * np.random.randn(100000,1)
m = X.shape[0]
y = 4 + 3*X + np.random.randn(100000,1)
X_b = np.c_[np.ones((m,1)),X]
theta = np.random.randn(X_b.shape[1],1)
```

```python
theta1 = np.random.randn(2,1)
```

```python
#    1 Iteration  - 3.13 Seconds         [ ]
#   10 Iteration  - 1.74 Milli Seconds   []
#  100 Iteration  - 3.26 Seconds         [4.0000653 , 2.99934014 ] 
# 1000 Iteration -  20.8 Ms              [4.0000653 , 2.99934014]
```

```python
%%time
n_iterations = 100
m = X_b.shape[0]
eta = 0.1

theta = theta1
theta_final = []
theta_final.append(theta.reshape(1,2))

for i in range(n_iterations):
    gradient = (2/m)*X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta *  gradient
    theta_final.append(theta.reshape(1,2))
print ("Value of theta:{}".format(theta))


plt.subplots(figsize=(20,12))
plt.scatter(X,y)
plt.plot(X_b[:,1],theta[0]+theta[1]*X_b[:,1],color='r')
for k in range(len(theta_final[:10])):
    #print (k)
    plt.plot(X_b[:,1],theta_final[k][:,0]+theta_final[k][:,1]*X_b[:,1],label=k)
    
#plt.plot(X_b[:,1],theta_final[1][:,0]+theta_final[1][:,1]*X_b[:,1],label='2')
#plt.plot(X_b[:,1],theta_final[2][:,0]+theta_final[2][:,1]*X_b[:,1],label='3')
plt.legend(loc = 'lower right ')
plt.xlabel('X')
plt.ylabel('y')
```
