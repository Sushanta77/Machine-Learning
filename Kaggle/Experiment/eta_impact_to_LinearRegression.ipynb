```python
from sklearn.datasets import fetch_mldata
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold,cross_val_score,cross_val_predict
from sklearn.metrics import confusion_matrix,precision_score,recall_score,precision_recall_curve
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mnist=fetch_mldata('MNIST Original')
```

```python
X=mnist['data']
y=mnist['target']
```

```python
X_train,y_train,X_test,y_test = X[:60000],y[:60000],X[60000:],y[60000:]
rand = np.random.permutation(60000)
X_train = X_train[rand]
y_train = y_train[rand]
```

```python
y_train_5 = (y_train == 5)
```

```python
sg = SGDClassifier(random_state=42)
```

```python
y_score = cross_val_predict(sg,X_train,y_train_5,cv=3,method='decision_function')
```

```python
precision,recall,thresholds = precision_recall_curve(y_train_5,y_score)
```

```python
plt.subplots(figsize=(16,8))
plt.plot(thresholds,precision[:-1],label='Precision')
plt.plot(thresholds,recall[:-1],label='Recall')
plt.xlim([-700000,700000])
plt.xlabel('Thresholds')
plt.legend(loc='lower right')
```

```python
n_iterations = 10000
m = 100
eta = 0.1

theta = np.random.randn(2,1)
theta_final = []
theta_final.append(theta.reshape(1,2))

for i in range(n_iterations):
    gradient = (2/m)*X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta *  gradient
    theta_final.append(theta.reshape(1,2))
print ("Value of theta:{}".format(theta))    
```

```python
plt.subplots(figsize=(20,12))
plt.scatter(X,y)
plt.plot(X_b[:,1],theta[0]+theta[1]*X_b[:,1],color='r')
for k in range(len(theta_final[:10])):
    #print (k)
    plt.plot(X_b[:,1],theta_final[k][:,0]+theta_final[k][:,1]*X_b[:,1],label=k)
    
#plt.plot(X_b[:,1],theta_final[1][:,0]+theta_final[1][:,1]*X_b[:,1],label='2')
#plt.plot(X_b[:,1],theta_final[2][:,0]+theta_final[2][:,1]*X_b[:,1],label='3')
plt.legend(loc = 'lower right ')
plt.xlabel('X')
plt.ylabel('y')
```
